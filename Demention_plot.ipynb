```python
import os,sys
import glob
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import umap
import matplotlib.pyplot as plt
```

```python
data_dir = '/mnt/data1/jlk/Cmap_database/CLUE/level3_cell_control'
data_3cell_file = 'MCF7_PC3_A549_landmark_random1000.npz'
data_6cell_file = 'MCF7_PC3_A549_A375_HT29_U2OS_landmark_random1000.npz'

def read_load_data(file):
    data = np.load(file,allow_pickle=True)
    x = data['x']
    y= data['y']
    return x,y

X,y = read_load_data('{}/{}'.format(data_dir,data_3cell_file))
print(X.shape,y.shape)
```

```python
plt.figure(figsize=(20,8))

######################
# demension plot： PCA
######################
X_PCA = PCA(n_components=2).fit_transform(X)
print(X_PCA.shape)

plt.subplot(131)
for i in range(3):
    plt.scatter(X_PCA[y==i,0],
                X_PCA[y==i,1],
                s=10,label = i )
plt.xticks([],[])
plt.yticks([],[])
plt.xlabel('PCA1')
plt.ylabel('PCA2')


######################
# demension plot： t-SNE
######################
X_tsne = TSNE(n_components=2, random_state=1).fit_transform(X)
print(X_tsne.shape)
plt.subplot(132)
for i in range(3):
    plt.scatter(X_tsne[y==i,0],
                X_tsne[y==i,1],
                s=10,label = i )
plt.xticks([],[])
plt.yticks([],[])
plt.xlabel('t-SNE1')
plt.ylabel('t-SNE2')


######################
# demension plot： UMAP
######################
X_umap = umap.UMAP(n_components=2).fit_transform(X)
print(X_umap.shape)
plt.subplot(133)
for i in range(3):
    plt.scatter(X_umap[y==i,0],
                X_umap[y==i,1],
                s=10,label = i )
plt.xticks([],[])
plt.yticks([],[])
plt.xlabel('Umap1')
plt.ylabel('Umap2')
plt.show()
```
